```python
import matplotlib.pyplot as plt
import random
from time import perf_counter_ns
time=[]
size=[8000,16000,32000,64000]
```

```python
def insertion_sort(A):
 
    for i in range(1,len(A)):
        pos = i-1   
        cur = A[i]
        while (A[pos] > cur and pos >=0):
            A[pos+1]=A[pos]
            pos = pos-1
        A[pos+1]=cur    
        i = i+1
    
    


    
for i in size:
    l = list(range(i))
    random.shuffle(l)
    start =  perf_counter_ns()
    insertion_sort(l)
    end = perf_counter_ns()
    time.append((end-start)*pow(10,-9))

```

```python
#----32K
print(time)
plt.plot(size,time,marker='o')
plt.xlabel("List Size")
plt.ylabel("Time in Seconds")
plt.show()
```
